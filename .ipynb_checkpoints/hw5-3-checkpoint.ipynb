```python
documents = [
        {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
        {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
        {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
      ]

directories = {
        '1': ['2207 876234', '11-2'],
        '2': ['10006'],
        '3': []
      }

def people(numbers):
  for doc_numbers in documents:
    if doc_numbers["number"] == numbers:
      print(doc_numbers["name"])
      break
  else:
    print('Такого номера документа нет в базе.')
def print_shelf(data, number_document):
    i=0
    for row in data.items():
        if number_document in row[1]:
            i+=1
            number_shelf = int (row[0])
            print('Документ лежит на полке №{:.0f}'.format(number_shelf))
    if i==0:
        print('Номер документа не правильный!')
def add_new_shelf(data, number_new_shelf):
    data.setdefault(number_new_shelf,[])
    directories=data
    print(directories)
def people_list():
  for persons in documents:
        for key, value in directories.items():
            if persons['number'] in value:
                print('№', persons["number"], 'Тип:', persons["type"], 'Владелец:' ,persons["name"],'Номер полки:', key)
def delete_polka(shelf_number):
    for k, v in directories.items():
        if shelf_number not in directories:
            print(f'Такой полки не существует. Текущий перечень полок:', ','.join(directories.keys()),'.')
        elif shelf_number and directories.get(shelf_number) != []:
            print(f'Полку нельзя удалить')
        else:
            shelf_number in directories and directories.get(shelf_number)==[]
            del directories[shelf_number]
            print(f'Полка удалена. Текущий перечень полок:', ','.join(directories.keys()),'.')
        break
def commands ():
    while True:
            command = input('\n \
              Введите одну из команд: p, l, s, ds, d. \n \
              Для выхода наберите exit. \n \
              Для вызов справки наберите help. \n \
              Ваша команда: ')
            if command == 'p':
                people(input('\nВведите номер документа:'))
            elif command == 'l':
                people_list()
            elif command == 's':
                number_document=(input('\nВведите номер документа:'))
                print_shelf(directories, number_document)
            elif command == 'ds':
                number_new_shelf=input('Введите номер новой полки')
                add_new_shelf(directories, number_new_shelf)
            elif command == 'd':
                shelf_number =input('Введите номер полки для удаления:')
                delete_polka(shelf_number)
            elif command == 'e':
                print('Вы завершили программу')
                break
            elif command == 'help':
                print('\n \
            p – команда спросит номер документа и выведет имя человека, которому он принадлежит;\n \
            l –  команда выведет список всех документов в формате passport "2207 876234" "Василий Гупкин";\n \
            s –  команда спросит номер документа и выведет номер полки, на которой он находится;\n \
            ds – команда добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться;\n \
            d – команда удаляет существующую полку из данных (только если она пустая).')

            elif command == 'q':
                print('Выход из программы')
            break
    else:
        print('Вы ввели некорректную команду, повторите ввод.')
commands ()
```
